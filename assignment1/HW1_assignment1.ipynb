```python
import pandas as pd
import numpy as np

# Read in data
da = pd.read_csv("HW1-data.csv")
da.head()
```

```python
# Calculate Average Rating and Reordering the data
#movie_list = list(da)[2:]
avg1 = da.drop(["User","Gender (1 =F, 0=M)"],axis=1).mean(axis=0).to_frame(name="avg_rating")
avg1 = avg1.sort_values(by="avg_rating",ascending=False)
avg1
```

```python
count1 = da.drop(["User","Gender (1 =F, 0=M)"],axis=1).count(axis=0).to_frame(name="total_count")
count1 = count1.sort_values(by="total_count",ascending=False)
count1
```

```python
# Count number 0f 4+ ratings
count2 = da[da >= 4].drop(["User","Gender (1 =F, 0=M)"],axis=1).count(axis=0).to_frame(name="high_rating")
count2 = pd.merge(count2,count1,left_index = True, right_index = True)
count2["high_rating_percent"] = count2["high_rating"] / count2["total_count"]
count2.sort_values(by="high_rating_percent",ascending = False)
```

```python
count3 = da[~pd.isnull(da["1: Toy Story (1995)"])].drop(["User","Gender (1 =F, 0=M)"],axis=1)\
    .count(axis=0).to_frame(name="cor_rating").reset_index()

count3["total_count"] = 17
count3["index_num"] = count3["index"].str.split(':').str[0].astype(int)
count3["cor_ratio"] = count3["cor_rating"] / count3["total_count"]
count3.sort_values(by=["cor_ratio","index_num"],ascending = [False,True])
```

```python
# Average Rating by gender
rating_by_gender = da.drop(["User"],axis = 1).groupby("Gender (1 =F, 0=M)").mean().T
rating_by_gender["diff"] = rating_by_gender[1] - rating_by_gender[0]
rating_by_gender.sort_values(by="diff",ascending = False)
```

```python
felmale_avg = da[da["Gender (1 =F, 0=M)"] == 1].drop(["User","Gender (1 =F, 0=M)"],axis = 1).stack().mean()
male_avg =  da[da["Gender (1 =F, 0=M)"] == 0].drop(["User","Gender (1 =F, 0=M)"],axis = 1).stack().mean()
print("Female Average Rating:{}, Male Average Rating:{}".format(felmale_avg,male_avg))
```

```python
# Get 4+ rating by gender
from IPython.display import display
def myCount(x):
    return np.sum(np.where(x>=4,1,0))
    
high_count = da.drop(["User"],axis=1).groupby("Gender (1 =F, 0=M)").agg([myCount]).T.reset_index()\
    .drop("level_1",axis = 1).rename(columns={"level_0" : "movie",1: "Female_High",0:"Male_High"})\
    .reset_index()[["movie","Male_High","Female_High"]]

count_by_gender = da.drop(["User"],axis = 1).groupby("Gender (1 =F, 0=M)").count().T\
    .reset_index().rename(columns = {"index":"movie",0: "Male_Count",1:"Female_Count"})

m1 = pd.merge(high_count,count_by_gender,left_on = "movie",right_on = "movie")
m1["Male_High_Percent"] = m1["Male_High"] / m1["Male_Count"]
m1["Female_High_Percent"] = m1["Female_High"] / m1["Female_Count"]
m1["diff"] = m1["Female_High_Percent"] - m1["Male_High_Percent"]
display(m1.sort_values(by="Male_High_Percent",ascending = False))
display(m1.sort_values(by="Female_High_Percent",ascending = False))
display(m1.sort_values(by="diff",ascending = False))
```

```python
print("Female High Rating:{}, Male High Rating:{}".format(sum(m1["Female_High"]) / sum(m1["Female_Count"]),
                                                                sum(m1["Male_High"]) / sum(m1["Male_Count"])))
```
