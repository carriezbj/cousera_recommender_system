```python
import pandas as pd
import numpy as np
da = pd.read_csv("data.csv",index_col = 0)

dummy = da.copy()
dummy[dummy > 0] = 1

uu_cor = da.T.corr()
uu_cor
```

```python
from IPython.display import display
print("Users Similar to 3867")
display(uu_cor.sort_values(by = 3867,ascending = False)[3867])
knn_3867 = uu_cor.sort_values(by = 3867,ascending = False)[3867].iloc[1:6]

print("Users Similar to 89")
display(uu_cor.sort_values(by = 89,ascending = False)[89])
knn_89 = uu_cor.sort_values(by = 89,ascending = False)[89].iloc[1:6]

print("Users Similar to 3712")
display(uu_cor.sort_values(by = 3712,ascending = False)[3712])
knn_3712 = uu_cor.sort_values(by = 3712,ascending = False)[3712].iloc[1:6]
```

```python
# Get weighted movie rating
raw_rating = np.dot(da.fillna(0).loc[knn_3867.index,:].T,knn_3867)
weights = np.dot(dummy.loc[knn_3867.index,:].fillna(0).T,knn_3867)
weights[weights == 0] = np.nan
fin = raw_rating / weights
rating1 = pd.DataFrame(fin,columns = ["3867"],index = list(da))

raw_rating = np.dot(da.fillna(0).loc[knn_89.index,:].T,knn_89)
weights = np.dot(dummy.loc[knn_89.index,:].fillna(0).T,knn_89)
weights[weights == 0] = np.nan
rating1["89"] = raw_rating / weights

raw_rating = np.dot(da.fillna(0).loc[knn_3712.index,:].T,knn_3712)
weights = np.dot(dummy.loc[knn_3712.index,:].fillna(0).T,knn_3712)
weights[weights == 0] = np.nan
rating1["3712"] = raw_rating / weights
```

```python
rating1.sort_values(by = "3867",ascending = False)
```

```python
rating1.sort_values(by = "89",ascending = False)
```

```python
# Normalized KNN
# Get r_u
avg_rating =  pd.DataFrame(da.mean(axis = 1),columns = ["avg_rating"],index = da.index)



# Get weighted movie rating
raw_rating = np.dot(da.loc[knn_3867.index,:].subtract(avg_rating.loc[knn_3867.index,:]["avg_rating"],axis = 0)\
                    .fillna(0).T,knn_3867)
weights = np.dot(dummy.loc[knn_3867.index,:].fillna(0).T,knn_3867)
weights[weights == 0] = np.nan
fin = raw_rating / weights + avg_rating.loc[3867,"avg_rating"]

rating2 = pd.DataFrame(fin,columns = ["3867"],index = list(da))

raw_rating = np.dot(da.loc[knn_89.index,:].subtract(avg_rating.loc[knn_89.index,:]["avg_rating"],axis = 0)\
                    .fillna(0).T,knn_89)

weights = np.dot(dummy.loc[knn_89.index,:].fillna(0).T,knn_89)
weights[weights == 0] = np.nan
rating2["89"] = raw_rating / weights + avg_rating.loc[89,"avg_rating"]

raw_rating = np.dot(da.loc[knn_3712.index,:].subtract(avg_rating.loc[knn_3712.index,:]["avg_rating"],axis = 0)\
                    .fillna(0).T,knn_3712)
weights = np.dot(dummy.loc[knn_3712.index,:].fillna(0).T,knn_3712)
weights[weights == 0] = np.nan
rating2["3712"] = raw_rating / weights + avg_rating.loc[3712,"avg_rating"]
```
