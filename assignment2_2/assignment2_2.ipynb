```python
import pandas as pd
import numpy as np
from IPython.display import display

da = pd.read_csv("data.csv",index_col = 0)
display(da)
display(da.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
s1 = pd.DataFrame(cosine_similarity(da.T.fillna(0)),index = list(da),columns = list(da))
s1.sort_values(by = "1: Toy Story (1995)",ascending = False)
```

```python
da_standardized = da.subtract(da.mean(axis = 1),axis = 0)

s2 = pd.DataFrame(cosine_similarity(da_standardized.T.fillna(0)),index = list(da),columns = list(da))
s2.sort_values(by = "1: Toy Story (1995)",ascending = False)
```

```python
# Get user recommendation
# Get rid of non-negative weights
s1[s1<0] = 0
s2[s2<0] = 0
# Get avg_rating per movie
movie_avg = da.mean(axis = 0).to_frame(name = "avg_rating")
ru = da.subtract(da.mean(axis = 0),axis = 1).loc[5277,:]

dummy = ru.copy()
dummy[dummy.notnull()] = 1

r1 = np.dot(s1,ru.fillna(0)) / np.dot(s1,dummy.fillna(0)) + da.mean(axis = 0)
r1.to_frame(name = "rating").sort_values(by = "rating",ascending = False)
```
