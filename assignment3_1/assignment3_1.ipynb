```python
import pandas as pd
import numpy as np
from IPython.display import display

act = pd.read_csv("act.csv",index_col = 0)
pred = pd.read_csv("pred.csv",index_col = 0)

diff = pred - act

display(act)
display(pred)
display(diff)
```

```python
# MAE by User
mae1 = diff.abs().mean(axis = 1).to_frame(name = "mae").reset_index()
mae1.sort_values(by = "mae",ascending = False)
```

```python
# MAE by movie
mae2 = diff.abs().mean(axis = 0).to_frame(name = "mae").reset_index()
mae2.sort_values(by = "mae",ascending = False)
```

```python
# Get MSE and RMSE per user
MSE1 = diff.applymap(lambda x: x**2).mean(axis = 1).to_frame(name="MSE").reset_index()
MSE1["RMSE"] = MSE1["MSE"].apply(np.sqrt)
display(MSE1.sort_values(by = "MSE",ascending = False))
```

```python
# Overall MSE
diff.applymap(lambda x: x**2).stack().mean()
```

```python
# Overall RMSE
np.sqrt(diff.applymap(lambda x: x**2).stack().mean())
```

```python
# Correlation between actual data and pred
act.corrwith(pred,axis = 1).to_frame("cor").reset_index().sort_values(by = "cor",ascending = False)
```

```python
# Get overall correlation
from scipy.stats import pearsonr
display(pearsonr(act.stack().values,pred.stack().values))
display(np.corrcoef(act.stack().values,pred.stack().values))
```
