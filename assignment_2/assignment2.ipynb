```python
import pandas as pd
import numpy as np
da = pd.read_csv("data2.csv",index_col = 0)
da
```

```python
# Get User Profiles
profiles = pd.DataFrame(np.dot(da.drop(["User_1","User_2","num-attr"],axis = 1).T,da[["User_1","User_2"]].fillna(0)),
             columns = ["profile_1","profile_2"],index = list(da.drop(["User_1","User_2","num-attr"],axis = 1)))
profiles
```

```python
# Get User Profiles
ratings = pd.DataFrame(np.dot(da.drop(["User_1","User_2","num-attr"],axis = 1),profiles),
                       columns = ["rating_1","rating_2"],index = da.index
             )
ratings.sort_values(by = "rating_1",ascending = False)
```

```python
ratings[ratings.rating_2 < 0].shape[0]
```

```python
##### Part II Standardize the weight ######
weighted_features = da.drop(["User_1","User_2","num-attr"],axis = 1).divide(np.sqrt(da["num-attr"]),axis = 0)
weighted_features
```

```python
# Get User Profiles
weighted_profiles = pd.DataFrame(np.dot(weighted_features.T,da[["User_1","User_2"]].fillna(0)),
             columns = ["profile_1","profile_2"],index = list(da.drop(["User_1","User_2","num-attr"],axis = 1)))
weighted_profiles
```

```python
ratings = pd.DataFrame(np.dot(weighted_features,weighted_profiles),
                       columns = ["rating_1","rating_2"],index = da.index
             )
ratings.sort_values(by = "rating_1",ascending = False)
```

```python
###### Part III Applying IDF ####
idf = da[da>0]
```

```python
IDF = 1/da[da>0].drop(["User_1","User_2","num-attr"],axis = 1).count()
```

```python
ratings = pd.DataFrame(np.dot(weighted_features*IDF,weighted_profiles),
                       columns = ["rating_1","rating_2"],index = da.index
             )
#ratings.sort_values(by = "rating_1",ascending = False)
ratings
```
